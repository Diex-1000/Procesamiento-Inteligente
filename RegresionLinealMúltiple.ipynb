```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Importe de Datos
df = pd.read_csv("./SalsberryRealty.csv")
df.head()
```

# 1. Determinar las variables independientes

```python
X = df[['Temp', 'Insul', 'Age']]
X
```

# 2. Determinar la variable dependiente

```python
y =df[['Cost']]
y
```

```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
```

# 3. Obtener el valor de intersecci√≥n

```python
model.intercept_
```

```python
print('Coefcientes: ', model.coef_)
print('Error cuadratico de la media', mse)
print('Variaza: ', r2)
```

```python
Y = model.predict([[30,5,10]])
print('El costo de calefaccion estimado es: ',Y)
```
